## Image To Biner

void biner(citra A, citra_biner B, int T, int M, int N)
/* Membuat citra biner dari citra A berdasarkan nilai ambang
(threshold) T yang dispesifikasikan. Ukuran citra adalah M
ï‚´ N.
citra_biner adalah tipe data untuk citra biner).
*/
{ int i, j;
citra_biner B;
for (i=0; i<=M-1; i++)
for (j=0; j<=N-1; j++)
{
if (A[i][j] < T)
B[i][j] = 0;
else
B[i][j] = 1;
}
}

```python
import numpy as np
import matplotlib.pyplot as plt

def biner(A, T):
    M, N = A.shape  # Mendapatkan ukuran citra
    B = np.zeros((M, N), dtype=int)  # Inisialisasi citra biner
    
    for i in range(M):
        for j in range(N):
            B[i, j] = 0 if A[i, j] < T else 1
    
    return B

# Membaca gambar dalam skala abu-abu menggunakan Matplotlib
image = plt.imread('Lena512warna.bmp')
if image.ndim == 3:
    image = np.dot(image[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB
T = 100
B = biner(image, T)

# Menampilkan hasil
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Citra Asli')

plt.subplot(1, 2, 2)
plt.imshow(B, cmap='gray')
plt.title('Citra Biner')

plt.show()
```

## Negative Image

void negatif(citra A, citra B, int M, int N)
/* Membuat citra negatif dari citra A. Hasilnya disimpan di
dalam citra B. Ukuran citra adalah M
ï‚´ N.
*/
{ int i, j;
for (i=0; i<=M-1; i++)
for (j=0; j<=N-1; j++)

```python
import numpy as np
import matplotlib.pyplot as plt

def negatif(A):
    M, N = A.shape  # Mendapatkan ukuran citra
    B = np.zeros((M, N), dtype=int)  # Inisialisasi citra negatif
    
    for i in range(M):
        for j in range(N):
            B[i, j] = 255 - A[i, j]
    
    return B

# Membaca gambar dalam skala abu-abu menggunakan Matplotlib
image = plt.imread('girl-warna.bmp')
if image.ndim == 3:
    image = np.dot(image[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB

B_negatif = negatif(image)

# Menampilkan hasil
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Citra Asli')

plt.subplot(1, 2, 2)
plt.imshow(B_negatif, cmap='gray')
plt.title('Citra Negatif')

plt.show()
```

## Image Brightening

void image_brightening(citra A, int b, citra B, int M, int N)
/* Pencerahan citra dengan menjumlahkan setiap pixel di dalam citra A dengan
sebuah skalar b. Hasil disimpan di dalam citra B. Citra berukuran M
ï‚´ N. */
{ int i, j, temp;
for (i=0; i<=M-1; i++)
for (j=0; j<=N-1; j++)
{
temp = A[i][j] + b;
/* clipping */
if (temp < 0)
B[i][j] = 0;
else
if (temp > 255)
B[i][j]=255;
else
B[i][j]=temp;
}
}

```python
import numpy as np
import matplotlib.pyplot as plt

def image_brightening(A, b):
    M, N = A.shape  # Mendapatkan ukuran citra
    B = np.zeros((M, N), dtype=int)  # Inisialisasi citra hasil
    
    for i in range(M):
        for j in range(N):
            temp = A[i, j] + b
            B[i, j] = max(0, min(255, temp))  # Clipping agar tetap dalam rentang [0, 255]
    
    return B

# Membaca gambar dalam skala abu-abu menggunakan Matplotlib
image = plt.imread('peppers512warna.bmp')
if image.ndim == 3:
    image = np.dot(image[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB

B_brightened = image_brightening(image, 50)

# Menampilkan hasil
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Citra Asli')

plt.subplot(1, 2, 2)
plt.imshow(B_brightened, cmap='gray')
plt.title('Citra Dicerahkan')

plt.show()
```

## Image Addition

void addition(citra A, citra B, citra C, int M, int N)
/* Menjumlahkan dua buah citra A dan B menjadi citra baru, C.
Citra A, B, dan C masing-masing berukuran M
ï‚´ N.
*/
{ int i, j, temp;
for (i=0; i<=M-1; i++)
for (j=0; j<=N-1; j++)
{
temp=A[i][j] + B[i][j];
if (temp > 255) C[i][j]=255; else C[i][j]=temp;
}
}

```python
import numpy as np
import matplotlib.pyplot as plt

def addition(A, B):
    M, N = A.shape  # Mendapatkan ukuran citra
    C = np.zeros((M, N), dtype=int)  # Inisialisasi citra hasil
    
    for i in range(M):
        for j in range(N):
            temp = A[i, j] + B[i, j]
            C[i, j] = min(255, temp)  # Clipping agar tidak melebihi 255
    
    return C

# Membaca gambar dalam skala abu-abu menggunakan Matplotlib
image1 = plt.imread('peppers512warna.bmp')
image2 = plt.imread('peppers512warna.bmp')

if image1.ndim == 3:
    image1 = np.dot(image1[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB
if image2.ndim == 3:
    image2 = np.dot(image2[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB

B_added = addition(image1, image2)

# Menampilkan hasil
plt.subplot(1, 3, 1)
plt.imshow(image1, cmap='gray')
plt.title('Citra 1')

plt.subplot(1, 3, 2)
plt.imshow(image2, cmap='gray')
plt.title('Citra 2')

plt.subplot(1, 3, 3)
plt.imshow(B_added, cmap='gray')
plt.title('Citra Hasil Penjumlahan')

plt.show()
```

## Image Substraction

void substraction (citra A, citra B, citra C, int M, int N)
/* Mengurangkan dua buah citra A dan B menajdi citra baru, C.
Citra A, B, dan C berukuran M
ï‚´ N.
*/
{ int i, j;
for (i=0; i<=M-1; i++)
for (j=0; j<=N-1; j++)
{
C[i][j]=A[i][j] - B[i][j];
if (C[i][j] != 0) C[i][j]=255; /* nyatakan objek berwarna putih */
}
}

```python
import numpy as np
import matplotlib.pyplot as plt

def substraction(A, B):
    M, N = A.shape  # Mendapatkan ukuran citra
    C = np.zeros((M, N), dtype=int)  # Inisialisasi citra hasil
    
    for i in range(M):
        for j in range(N):
            C[i, j] = A[i, j] - B[i, j]
            if C[i, j] != 0:
                C[i, j] = 255  # Nyatakan objek berwarna putih
    
    return C

# Membaca gambar dalam skala abu-abu menggunakan Matplotlib
image1 = plt.imread('Lena512warna.bmp')
image2 = plt.imread('Lena512warna.bmp')

if image1.ndim == 3:
    image1 = np.dot(image1[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB
if image2.ndim == 3:
    image2 = np.dot(image2[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB

B_subtracted = substraction(image1, image2)

# Menampilkan hasil
plt.subplot(1, 3, 1)
plt.imshow(image1, cmap='gray')
plt.title('Citra 1')

plt.subplot(1, 3, 2)
plt.imshow(image2, cmap='gray')
plt.title('Citra 2')

plt.subplot(1, 3, 3)
plt.imshow(B_subtracted, cmap='gray')
plt.title('Citra Hasil Pengurangan')

plt.show()
```

## Image Multiplication

void multiplication(citra A, citra B, citra C, int M, int N)
/* Mengalikan citra A dengan citra B menjadi citra C.
Citra A, matriks B, dan hasil perkalian C berukuran M
ï‚´ N.
*/
{ int i, j;
for (i=0; i<=M-1; i++)
for (j=0; j<=N-1; j++)
for (k=0; k<=N-1; k++)
{ C[i][j] = A[i][j]*B[i][j];
/* clipping */
if (C[i][j] < 0)
C[i][j] = 0;
else
if (C[i][j] > 255)
C[i][j]=255;
}
}

```python
import numpy as np
import matplotlib.pyplot as plt

def multiplication(A, B):
    M, N = A.shape  # Mendapatkan ukuran citra
    C = np.zeros((M, N), dtype=int)  # Inisialisasi citra hasil
    
    for i in range(M):
        for j in range(N):
            C[i, j] = A[i, j] * B[i, j]
            C[i, j] = max(0, min(255, C[i, j]))  # Clipping agar tetap dalam rentang [0, 255]
    
    return C

# Membaca gambar dalam skala abu-abu menggunakan Matplotlib
image1 = plt.imread('baboon24.bmp')
image2 = plt.imread('baboon24.bmp')

if image1.ndim == 3:
    image1 = np.dot(image1[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB
if image2.ndim == 3:
    image2 = np.dot(image2[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB

B_multiplied = multiplication(image1, image2)

# Menampilkan hasil
plt.subplot(1, 3, 1)
plt.imshow(image1, cmap='gray')
plt.title('Citra 1')

plt.subplot(1, 3, 2)
plt.imshow(image2, cmap='gray')
plt.title('Citra 2')

plt.subplot(1, 3, 3)
plt.imshow(B_multiplied, cmap='gray')
plt.title('Citra Hasil Perkalian')

plt.show()
```

## Boolean Operation NOT

void not(citra_biner A, citra_biner B, int M, int N)
/* Membuat citra komplemen dari citra biner A.
Komplemennya disimpan di dalam B. Ukuran citra A
adalah M
ï‚´ N.
*/
{ int i, j;
for (i=0; i<=M-1; i++)
for (j=0; j<=N-1; j++)
{
B[i][j] = !A[i][j];
}
}

```python
import numpy as np
import matplotlib.pyplot as plt

def binary_not(A):
    M, N = A.shape  # Mendapatkan ukuran citra
    B = np.zeros((M, N), dtype=int)  # Inisialisasi citra hasil
    
    for i in range(M):
        for j in range(N):
            B[i, j] = 1 - A[i, j]  # Operasi NOT pada citra biner
    
    return B

# Membaca gambar dalam skala abu-abu menggunakan Matplotlib
image = plt.imread('peppers512warna.bmp')

if image.ndim == 3:
    image = np.dot(image[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB

# Konversi ke citra biner
threshold = 128
binary_image = (image >= threshold).astype(int)

B_complemented = binary_not(binary_image)

# Menampilkan hasil
plt.subplot(1, 2, 1)
plt.imshow(binary_image, cmap='gray')
plt.title('Citra Biner Asli')

plt.subplot(1, 2, 2)
plt.imshow(B_complemented, cmap='gray')
plt.title('Citra Biner Komplemen')

plt.show()
```

## Image Translation

void translation(citra A, citra B, int M, int N, int m, int n)
/* Mentranslasi citra A sejauh m, n menjadi citra B. Ukuran citra M
ï‚´ N. */
{ int i, j;
for (i=0; i<=M-1; i++)`
for (j=0; j<=N-1; j++)
{
B[i][j]=A[i+m][j+n];
}
}

```python
import numpy as np
import matplotlib.pyplot as plt

def translation(A, m, n):
    M, N = A.shape  # Mendapatkan ukuran citra
    B = np.zeros((M, N), dtype=A.dtype)  # Inisialisasi citra hasil
    
    for i in range(M):
        for j in range(N):
            if 0 <= i + m < M and 0 <= j + n < N:
                B[i + m, j + n] = A[i, j]  # Translasi citra
    
    return B

# Membaca gambar dalam skala abu-abu menggunakan Matplotlib
image = plt.imread('girl-warna.bmp')

if image.ndim == 3:
    image = np.dot(image[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB

# Parameter translasi
m_shift = 30  # Geser ke bawah
n_shift = 50  # Geser ke kanan

B_translated = translation(image, m_shift, n_shift)

# Menampilkan hasil
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Citra Asli')

plt.subplot(1, 2, 2)
plt.imshow(B_translated, cmap='gray')
plt.title('Citra Hasil Translasi')

plt.show()
```

## Image Rotation Counter Clock-wise

void rotation90CCW(citra A, citra B, int M, int N)
/* Rotasi citra A sejauh 90
ï‚° berlawanan arah jarum jam (CCW = Clock Counter-
wise). Ukuran citra adalah M
ï‚´ N. Hasil rotasi disimpan di dalam citra B.
*/
{ int i, j, k;
for (i=0; i<=M-1; i++)
{
k=N-1;
for (j=0; j<=N-1; j++)
{
B[k][i]=A[i][j];
k--;
}
}
}

```python
import numpy as np
import matplotlib.pyplot as plt

def rotation90CCW(A):
    M, N = A.shape  # Mendapatkan ukuran citra
    B = np.zeros((N, M), dtype=A.dtype)  # Inisialisasi citra hasil dengan ukuran terbalik
    
    for i in range(M):
        k = N - 1
        for j in range(N):
            B[k, i] = A[i, j]
            k -= 1
    
    return B

# Membaca gambar dalam skala abu-abu menggunakan Matplotlib
image = plt.imread('Lena512warna.bmp')

if image.ndim == 3:
    image = np.dot(image[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB

B_rotated = rotation90CCW(image)

# Menampilkan hasil
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Citra Asli')

plt.subplot(1, 2, 2)
plt.imshow(B_rotated, cmap='gray')
plt.title('Citra Hasil Rotasi 90Â° CCW')

plt.show()
```

## Image Flipping

void vertical_flip(citra A, citra B, int M, int N)
/* Flipping vertikal (pencerminan terhadap sumbu-X) terhadap citar A. */
Ukuran citra adalah M
ï‚´ N. Hasil flipping disimpan di dalam citra B.
*/
{ int i, j, k;
k=N-1;
for (i=0; i<=M-1; i++)
{
for (j=0; j<=N-1; j++)
{
B[k][j]=A[i][j];
}
k--;
}
}

```python
import numpy as np
import matplotlib.pyplot as plt

def vertical_flip(A):
    M, N = A.shape  # Mendapatkan ukuran citra
    B = np.zeros((M, N), dtype=A.dtype)  # Inisialisasi citra hasil
    
    k = M - 1
    for i in range(M):
        for j in range(N):
            B[k, j] = A[i, j]
        k -= 1
    
    return B

# Membaca gambar dalam skala abu-abu menggunakan Matplotlib
image = plt.imread('Lena512warna.bmp')

if image.ndim == 3:
    image = np.dot(image[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB

B_flipped = vertical_flip(image)

# Menampilkan hasil
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Citra Asli')

plt.subplot(1, 2, 2)
plt.imshow(B_flipped, cmap='gray')
plt.title('Citra Hasil Flipping Vertikal')

plt.show()
```

## Image Zooming

void zoom_out(citra A, citra B, int M, int N)
/* perbesaran citra A dengan faktor skala 2
Ukuran citra adalah M
ï‚´ N. Hasil perbesaran disimpa d dalam citra B.
*/
{ int i, j, k, m, n;
m=0; n=0;
for (i=0; i<=M-1; i++)
{
for (j=0; j<=N-1; j++)
{
B[m][n]= A[i][j];
B[m][n+1]= A[i][j];
B[m+1][n]= A[i][j];
B[m+1][n+1]= A[i][j];
n=n+2;
}
m=m+2;
n=0;
}
}

```python
import numpy as np
import matplotlib.pyplot as plt

def zoom_out(A):
    M, N = A.shape  # Mendapatkan ukuran citra
    B = np.zeros((M * 2, N * 2), dtype=A.dtype)  # Inisialisasi citra hasil dengan ukuran diperbesar
    
    m, n = 0, 0
    for i in range(M):
        for j in range(N):
            B[m, n] = A[i, j]
            B[m, n + 1] = A[i, j]
            B[m + 1, n] = A[i, j]
            B[m + 1, n + 1] = A[i, j]
            n += 2
        m += 2
        n = 0
    
    return B

# Membaca gambar dalam skala abu-abu menggunakan Matplotlib
image = plt.imread('baboon24.bmp')

if image.ndim == 3:
    image = np.dot(image[..., :3], [0.2989, 0.587, 0.114])  # Konversi ke grayscale jika RGB

B_zoomed = zoom_out(image)

# Menampilkan hasil
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Citra Asli')

plt.subplot(1, 2, 2)
plt.imshow(B_zoomed, cmap='gray')
plt.title('Citra Hasil Perbesaran')

plt.show()
```
